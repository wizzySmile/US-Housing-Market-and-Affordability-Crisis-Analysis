# Data Cleaning for United States CPI_data from 2015-2024
- Droped unnecessary index
- Reset index
- Change column names
- Removed unnecessary columns
- Data standardaziation
- Check for data consistency


```python
cpi_data = pd.read_csv("D:/project folder/US-Housing-Market-and-Affordability-Crisis-Project/dataset/raw-data/cpi_data.csv")
cpi_data
```

```python
# Rename columns
cpi_data.columns = ['Date', 'Average_CPI']

# Convert to datetime
cpi_data['Date'] = pd.to_datetime(cpi_data['Date'])
```

```python
cpi_data = cpi_data[cpi_data['Date'] >= '2018-01-01']
cpi_data
```

```python
# Set 2018-01-01 as base month
base_month = '2018-01-01'
base_cpi = cpi_data[cpi_data['Date'] == base_month]['Average_CPI'].values[0]

# Calculate CPI Index
cpi_data.loc[:, 'CPI_Index'] = (cpi_data.loc[:,'Average_CPI'] / base_cpi) * 100

# Keep relevant columns
monthly_cpi = cpi_data[['Date', 'Average_CPI', 'CPI_Index']]

monthly_cpi
```

```python
monthly_cpi = monthly_cpi.reset_index(drop=True)
monthly_cpi.head()
```

```python
monthly_cpi = monthly_cpi.copy()
monthly_cpi.loc[:, 'Year'] = monthly_cpi['Date'].astype(str).str[:4].astype(int)
```

```python
cpi_2018_2024 = monthly_cpi
```

```python
cpi_2018_2024 = cpi_2018_2024[['Year', 'Average_CPI', 'CPI_Index']]
cpi_2018_2024
```

```python
cpi_df = cpi_2018_2024.groupby('Year').agg({'Average_CPI': 'mean'}).reset_index()
cpi_df
```

```python
# Set 2018 as base year
base_year = 2018
base_cpi = cpi_df[cpi_df['Year'] == base_year]['Average_CPI'].values[0]

# Calculate CPI Index
cpi_df.loc[:, 'CPI_Index'] = (cpi_df['Average_CPI'] / base_cpi) * 100

# Keep relevant columns
yearly_cpi = cpi_df[['Year', 'Average_CPI', 'CPI_Index']]

yearly_cpi
```

```python
# save monthly_cpi as cpi_2018_2024.csv
yearly_cpi.to_csv('cpi_2018_2024.csv', index=False)
```

# Data Cleaning for United States house estimate data from 2015-2024
- Droped unnecessary index
- Reset index
- Change column names
- Removed unnecessary columns
- Data standardaziation
- Check for data consistency

```python
import requests

years = {
    2018:"acs1",
    2019:"acs1",
    2020:"acs5",
    2021:"acs1",
    2022:"acs1",
    2023:"acs1",
    2024:"acs1"
}

all_data = []

for year, dataset in years.items():
    url = f"https://api.census.gov/data/{year}/acs/{dataset}?get=NAME,B11001_001E&for=state:*"
    r = requests.get(url).json()

    df = pd.DataFrame(r[1:], columns=r[0])
    df["Year"] = year
    all_data.append(df)

household_estimate = pd.concat(all_data)
household_estimate.head(3)
```

```python
household_estimate = household_estimate[~household_estimate['NAME'].isin(['District of Columbia', 'Puerto Rico'])]
```

```python
household_estimate[household_estimate['NAME'].isin(['District of Columbia', 'Puerto Rico'])]
```

```python
household_estimate = household_estimate.rename(columns= {'NAME': 'State_Name', 
                           'B11001_001E': 'Household_Estimate'})
household_estimate.head(3)
```

```python
household_estimate = household_estimate.drop('state', axis=1)
household_estimate.head(3)
```

```python
# household_estimate as household_estimate_2018_2024
household_estimate.to_csv('household_estimate_2018_2024.csv', index=False)
```
