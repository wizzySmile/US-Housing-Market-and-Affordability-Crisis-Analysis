### Create Growth Metrics (Within Each State)
Creating columns that calculate growth per state..
- Home Value Growth (YoY %)
- Income Growth (YoY %)

```python

affordability_full_2018_2024 = affordability_full_2018_2024.sort_values(["State_Name", "Year"])

affordability_full_2018_2024 = affordability_full_2018_2024.assign(Home_Value_YoY_pct = (
    affordability_full_2018_2024.groupby("State_Name")["Median_Home_Value"]
      .pct_change() * 100),
                                                                   Income_YoY_pct = (
    affordability_full_2018_2024.groupby("State_Name")["Median_Income"]
      .pct_change() * 100)                   
                                                                  )
```

```python
affordability_full_2018_2024
```

### Why This Matters

This allows for direct comparison:

Is home price growth > income growth?

***Note:*** That’s the first real affordability signal.

-----------------------------------------------------------------------------------------------------------------------------------

### Inflation Adjustment (Real Values)
Creating columns that convert nominal values to real values by adjusting for inflation
- Real_Home_Value
- Real_Income

```python
affordability_full_2018_2024 = (affordability_full_2018_2024
                                .assign(Real_Home_Value = (affordability_full_2018_2024["Median_Home_Value"] / 
                                        affordability_full_2018_2024["CPI_Index"] * 100),
                                        Real_Income = (affordability_full_2018_2024["Median_Income"] / 
                                                       affordability_full_2018_2024["CPI_Index"] * 100)))
```

```python
affordability_full_2018_2024.head(2)
```

### Why This Matters

If home values rose 10% but inflation rose 8%, real growth is only 2%.

This prevents false crisis signals.

-----------------------------------------------------------------------------------------------------------------------------------

### Mortgage Burden (Critical Step)

compute true affordability pressure.

Estimated monthly mortgage payment using amortization formula.

```python

r = affordability_full_2018_2024["Mortgage_Rate_pct"] / 100 / 12
n = 30 * 12

affordability_full_2018_2024["Monthly_Mortgage"] = (
    affordability_full_2018_2024["Median_Home_Value"] * 0.8 *
    (r * (1 + r)**n) / ((1 + r)**n - 1)
)
```

```python
affordability_full_2018_2024["Mortgage_Burden_pct"] = (
    affordability_full_2018_2024["Monthly_Mortgage"] * 12 / affordability_full_2018_2024["Median_Income"]
) * 100
```

```python
affordability_full_2018_2024[affordability_full_2018_2024['Mortgage_Burden_pct']>=30]
```

### What it means 
If Mortgage_Burden_pct = 35%

It means:

35% of annual income is required to service the mortgage.

Above 30% is generally considered financially stressed.

- Below 20% → very affordable states/years
- 30%+ → affordability stress
- 50%+ → severe pressure

-----------------------------------------------------------------------------------------------------------------------------------

### Rent Burden %

Calculate rental pressure.

```python
affordability_full_2018_2024["Rent_Burden_pct"] = (
    affordability_full_2018_2024["Median_Rent_Price"] * 12 / affordability_full_2018_2024["Median_Income"]
) * 100
```

```python
affordability_full_2018_2024.head(3)
```

### Now we can compare:

Buying burden vs Renting burden

Which is worse?

Is renting becoming equally unaffordable?

### Affordability Gap and Housing Cost
This metric measures the divergence between housing costs and affordability, indicating how accessible houses are to potential buyers.

```python
affordability_full_2018_2024['Annual_Housing_Cost'] = (
    affordability_full_2018_2024['Monthly_Mortgage'] * 12 +
    (affordability_full_2018_2024['Median_Home_Value'] * 0.012) +   # property tax
    (affordability_full_2018_2024['Median_Home_Value'] * 0.01)      # maintenance
)
```

```python
affordability_full_2018_2024['Affordability_pct'] = (
    affordability_full_2018_2024['Annual_Housing_Cost'] / affordability_full_2018_2024['Median_Income']
) * 100
```

```python
# Required_Income = Annual Housing Cost / Affordability Threshold(30%)
affordability_full_2018_2024['Required_Income'] = affordability_full_2018_2024['Annual_Housing_Cost'] / 0.3
```

```python
affordability_full_2018_2024["Affordability_Gap_pct"] = (
    (affordability_full_2018_2024["Required_Income"] - affordability_full_2018_2024["Median_Income"])
    / 
    affordability_full_2018_2024["Required_Income"] * 100
)
```

```python
affordability_full_2018_2024.head(2)
```

### It Means..
If positive → housing is outpacing income.

If negative → income catching up.
